## Load Metadata and Get Repeated Samples

```python
# load metadata and index by Data File ID (name of real file)
metadata = pd.read_csv(
    "/home/yraghav/Joint-Genotyping-Fall-2021/inputs/metadata/ALS_Consortium_WGS_Metadata.csv").set_index(
        "Data File ID"
    )
```

```python
# value counts of number of subjects
num_times = metadata["ExternalSubjectId"].value_counts()

# get IDs for patients who have been sequenced more than once
repeated_patients = num_times[num_times > 1].index
```

```python
repeated_patients
```

## Choose Between Repeated Samples

```python
# lists for selected and removed samples 
selected = []
removed = []

# for each patient that's been repeated 
for patient in repeated_patients:
    
    # subset to patient 
    patient_df = metadata[metadata["ExternalSubjectId"]==patient]
    
    # get the newest quote for the repeated subject
    newest_quote = patient_df["Quote"].sort_values(ascending=False)[0]
    
    # get list of samples in the newest quote 
    candidates = patient_df[patient_df["Quote"]==newest_quote]
    
    # if the patient was sequenced once in the newest quote (ideal)
    if candidates.index.size == 1: 
                
        # append selected sample
        selected.append(candidates.index[0])
        
        # append removed samples 
        removed.extend(
            patient_df[patient_df.index!=candidates.index[0]].index
        )
        
    # choose between samples in the newest quote
    else:
        
        candidates 
        
        # if there are blood samples, take them 
        blood_candidates = candidates[
            candidates["Sample Tissue Source"] == "Blood"
        ]
        
        # if there are blood samples, just take first of these samples (random). 
        if blood_candidates.index.size != 0: 
        
            selected.append(blood_candidates.index[0])
            
            removed.extend(
                patient_df[patient_df.index!=blood_candidates.index[0]].index
            )

        # if no blood samples, just take first sample (random)
        else: 
            
            selected.append(candidates.index[0])
            
            removed.extend(
                patient_df[patient_df.index!=candidates.index[0]].index
            )
            
            
    
    
```

## Write Out Selected & Removed Files

```python
with open ("../output/selected.txt", 'w') as out_file: 
    for id in selected: 
        _ = out_file.write(id + "\n")
```

```python
with open ("../output/removed.txt", 'w') as out_file: 
    for id in removed: 
        _ = out_file.write(id + "\n")
```

## Create Remove Data Script

```python
with open("../output/remove_repeated_samples_data.sh", 'w') as out_file: 
    
    _ = out_file.write("# Script MUST be run on C3DDB\n")
    
    for id in removed: 
        _ = out_file.write("rm -f /scratch/users/yraghav/JG_2021_VCF_data/*{}*\n".format(id))
```
